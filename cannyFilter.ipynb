```python
%matplotlib inline
import imageio
from skimage import feature
import matplotlib.pyplot as plt
import numpy as np
```

```python
def loadImage(file_name):
    """ PURPOSE: Load image """

    return imageio.imread(file_name, as_gray=True)

def createEdges(image, sigma_opt_1, sigma_opt_2, sigma_opt_3):
    """ PURPOSE: Generate edges through Canny Filter  """

    edges_1 = feature.canny(image, sigma_opt_1)
    edges_2 = feature.canny(image, sigma_opt_2)
    edges_3 = feature.canny(image, sigma_opt_3)
    
    return edges_1, edges_2, edges_3
```

```python
if __name__ == '__main__':

    gif_storm = loadImage("frame_000_delay-0.04s.png")

    storm_edges_1, storm_edges_2, storm_edges_3 = createEdges(gif_storm, 1, 3, 5)
    
    # Gaussian Filter
    plt.imshow(storm_edges_1, cmap=plt.cm.gray)
    plt.title("Canny Filter on Gif Storm | Sigma 1")
    plt.show()
    
    plt.imshow(storm_edges_2, cmap=plt.cm.gray)
    plt.title("Canny Filter on Gif Storm | Sigma 3")
    plt.show()
    
    plt.imshow(storm_edges_3, cmap=plt.cm.gray)
    plt.title("Canny Filter on Gif Storm | Sigma 5")
    plt.show()
```
